### Reindex

For Row:

```python
sales_jan = pd.DataFrame({
    'Sales': [250, 400, 300]
}, index=['Store_A', 'Store_B', 'Store_C'])

sales_jan
```

```python
sales_feb = pd.DataFrame({
    'Sales': [260, 310]
}, index=['Store_A', 'Store_C'])

sales_feb
```

```python
sales_feb - sales_jan
```

```python
sales_feb_aligned = sales_feb.reindex(sales_jan.index)

sales_feb_aligned
```

```python
sales_feb_aligned = sales_feb.reindex(sales_jan.index, fill_value=0)

sales_feb_aligned
```

```python
sales_feb_aligned - sales_jan
```

For Column:

```python
source_a = pd.DataFrame({
    'Product': ['Laptop', 'Tablet', 'Phone'],
    'Price': [1000, 500, 800],
    'Stock': [30, 50, 100]
})

source_a
```

```python
source_b = pd.DataFrame({
    'Price': [980, 520, 810],
    'Product': ['Laptop', 'Tablet', 'Phone']
})

source_b
```

```python
source_a.columns
```

```python
source_b_aligned = source_b.reindex(columns=source_a.columns, fill_value=0)

source_b_aligned
```

### Iteration in Pandas

```python
df = pd.DataFrame({
    'Name' : ['heartine', 'gagan', 'gavin'],
    'Age' : [25, 20, 23],
    'City' : ['Delhi', 'Mumbai', 'Chennai'],
})
```

```python
df
```

```python
for index, row in df.iterrows():
    print (f"{index}. {row['Name']}")
    print ("="*15)
```

```python
for row in df.itertuples():
    print (f"{row.Name}")
    print ("="*15)
```

```python
for col in df.columns:
    print(f"Column: {col}")
    print (df[col].values)
    print ("="*15)
```

### Condition based Columns Creation

```python
df
```

```python
categories = []

for index, row in df.iterrows():
    if row["Age"] > 22:
        categories.append("Senior")
    else:
        categories.append("Junior")

categories
```

```python
df["Category"] = categories
```

```python
df
```

```python
df["Maturity"] = df["Age"].apply(lambda age : "Yes" if age > 22 else "No")
```

```python
df
```
